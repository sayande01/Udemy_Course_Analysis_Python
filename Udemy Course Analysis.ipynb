# Udemy Course Data Analysis

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
import plotly.express as px
warnings.filterwarnings('ignore')
```

```python
sns.set()
```

```python
df= pd.read_csv(r'udemy.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
#No of Courses
```

```python
df['course_id'].nunique()
```

```python
# change date-time column data type Object to DateTime 
```

```python
df['published_timestamp'] = pd.to_datetime(df['published_timestamp'])
```

```python
df.info()
```

```python
# Creating a Year Column
```

```python
df['Year'] = df['published_timestamp'].dt.year
```

```python
df
```

```python
# Year wise No of Courses Publised
```

```python
df1 = df.groupby('Year')['course_id'].count()
df1
```

```python
sns.barplot(x = df1.index, y = df1.values, palette = 'dark')
plt.xlabel('Year')
plt.ylabel('No of Courses')
plt.title('No of Course price by Year')
plt.show()
```

```python
df6 = df.groupby('Year')['num_subscribers'].sum()
df6
```

```python
ax = sns.lineplot(x = df6.index, y = df6.values)
ax.ticklabel_format(style='plain', axis='y')
plt.xlabel('Year')
plt.ylabel('No of Subscribers')
plt.title('No of Subscribers by Year')
plt.show()
```

```python
# deleting the entries with price = free
```

```python
df2 = df[~(df['price']== 'Free')]
df2
```

```python
df2[(df2['price']== 'Free')]
```

```python
# in df2 all 'free' priced are deleted
```

```python
df2['price']= df2['price'].astype(int)
```

```python
# Yearly avg of course price
```

```python
df3 = df2.groupby('Year')['price'].mean().astype(int)
df3
```

```python
sns.lineplot(x = df3.index, y = df3.values)
plt.xlabel('Year')
plt.ylabel('Avg Course Price')
plt.title('Average of Course price by Year')
plt.show()
```

```python
# Level wise distribution of courses
```

```python
df4 = df2['level'].value_counts()
df4
```

```python
df4.plot(kind = 'pie', autopct = '%1.2f%%')
plt.title('Distribution of no of courses wrt Level')
plt.show()
```

```python
# Top 5 Courses
```

```python
df2[['course_title','num_subscribers']].head().sort_values(by = 'num_subscribers', ascending = False)
```

```python
# No of Course by Subject
```

```python
df5 = df['subject'].value_counts()
df5
```

```python
sns.barplot(x = df5.values, y = df5.index, palette = 'husl')
plt.xlabel('Year')
plt.ylabel('No of Course')
plt.title('No of Course by Subject')
plt.show()
```

```python
df_grouped = df2.groupby('subject')['num_lectures'].sum()
df_grouped
```

```python

plt.figure(figsize=(10, 6)) 
df_grouped.plot.area()

# Adding labels and title
plt.xlabel('Subject')
plt.ylabel('Total Number of Lectures')
plt.title('Total Number of Lectures by Subject')

plt.show()
```

```python
df7 = df2.groupby('subject')['num_reviews'].sum()
df7
```

```python


labels = df7.index
sizes = df7.values

# Create a donut chart
plt.figure(figsize=(6, 6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, colors=['skyblue', 'lightgreen', 'Pink','blue'], wedgeprops=dict(width=0.4))
plt.title('Subject wise Reviews distribution')
plt.axis('equal')
plt.show()
```

```python
# Dustribution of Price Subject wise
```

```python
#Histogram
plt.figure(figsize=(8, 6))
sns.histplot(data = df2, x='price', bins=15, kde=True, hue = 'subject',palette = 'dark')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.title('Histogram of Price')
plt.show()
```

```python
# Heatmap of Price
```

```python

pivot_table = df2.pivot_table(index='Year', columns='subject', values='price', aggfunc=np.mean)
plt.figure(figsize=(10, 6))
sns.heatmap(pivot_table, cmap='Reds', annot=True, linewidths=0.5)
plt.xlabel('Subject')
plt.ylabel('Year')
plt.title('Heatmap of Avg Price of courses by Year for each Subjects')
plt.show()
```
